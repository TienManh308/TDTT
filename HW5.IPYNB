```python
#3
def Is_Prime(a):
    if a < 2:
        return False
    if a < 4:
        return True
    if a % 2 == 0 or a % 3 == 0:
        return False
    i = 5
    while i * i <= a:
        if a % i == 0 or a % (i+2) == 0:
            return False
        i+=6
    return True

Is_Prime(5)
```

```python
#4
def is_perfect_number(n):
    if n <= 0:
        return False
        
    sum = 0
    
    for i in range(1, n):
        if n % i == 0:
            sum += i      
    return sum == n

data = [23, 6, 28, 123, 24]

print("Kiểm tra Số Hoàn Hảo:")
for num in data:
    is_perfect = is_perfect_number(num)
    print(f"Số {num}: {is_perfect}")
```

```python
#5
def vt(n, a):
    for i in range (len(a)):
        if n == a[i]:
            return i + 1
    return -1

li = list(map(int, input().split()))
num = int(input())

vt(num, li)
```

```python
#6
def giai_thua(n):
    if n == 1: return 1
    return giai_thua(n-1)*n

giai_thua(4)
```

```python
#7
def may_tinh(st):
    li = st.split()
    if li[1] == '+': return round(float(li[0]) + float(li[2]),2)
    if li[1] == '-': return round(float(li[0]) - float(li[2]),2)
    if li[1] == '*': return round(float(li[0]) * float(li[2]),2)
    return round(float(li[0]) / float(li[2]),2)

s = input()
may_tinh(s)
```

```python
#8
#8
def hamming_distance(num1, num2):
    cnt = 0
    while num1 or num2:
        cnt += (num1 & 1) ^ (num2 & 1)
        num1 >>= 1
        num2 >>= 1
    return cnt

a, b = map(int,input().split())

print(hamming_distance(a, b))
```

```python
#9
def Sum(s):
    sumarize = 0
    for i in range(len(s)):
        sumarize += int(s[i])

    return sumarize

n = input()
Sum(n)
```

```python

#10
def check(str1, str2):
    func = {}
    for i in range(0, len(str1)):
        if func.get(str1[i]) == None:
            func[str1[i]] = str2[i]
        elif func.get(str1[i]) != str2[i]:
            return False
    return True

a, b = input().split()

if check(a, b) and check(b, a):
    print('true')
else:
    print('false')
```
