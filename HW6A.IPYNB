```python
#1
li = list(map(int, input().split()))

dic = {} 
sol = [] 

for item in li:
    if item not in dic:
        dic[item] = True
        sol.append(item)

print(sol)
```

```python
#2
li = list(map(int, input().split()))
s = []
s.append(li[0])
for i in range(1, len(li)):
    s.append(s[i-1] + li[i])
s
```

```python
#3
tup = tuple(map(int, input().split()))
k = int(input())
k = k % len(tup)
tup = tup[k:]+ tup[:k]
tup
```

```python
#4
li = input().split()
dic = dict()

for i in range(len(li)):
    t = li[i].split(':')
    # Lấy key = t[0], value = t[1]
    key = t[0]
    value = t[1]
    
    # Nếu key chưa có, dic.get sẽ trả về [], rồi append value vào
    dic[key] = dic.get(key, []) + [value]  # Dùng [value] để thêm vào list

print(dic)
```

```python
#5
li = list(map(int, input().split()))
dic = {'positive': 0, 'negative': 0, 'zero': 0}
for i in range(len(li)):
    if li[i] == 0:
        dic['zero'] += 1
    elif li[i] < 0:
        dic['negative'] += 1
    else:
        dic['positive'] += 1

dic
```
