```python
#1
def binary_search(arr, left, right, val):
    mid = (left+right)//2
    if arr[mid] < val: 
        return (binary_search(arr, mid + 1, right, val))
    elif arr[mid] > val:
        return(binary_search(arr, left, mid + 1, val))
    elif arr[mid] == val:
        return mid
    return -1

lst = [int(n) for n in input().split()]
k = int(input())
print(binary_search(lst, 0, len(lst) - 1,k))
                        
```

```python
#3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1): 
        for j in range(n - i - 1): 
            if arr[j] > arr[j+1]: 
                arr[j],arr[j+1] = arr[j+1], arr[j]
    return arr

def merge_sort(arr, left, right):
    if len(arr) <= 1: 
        return arr
    mid = (left+right)//2
    left_arr = arr[:mid]
    right_arr = arr[mid:]
    a = merge_sort(left_arr,0, mid)
    b = merge_sort(right_arr, mid, len(arr)- 1)
    return merge(a,b)

def merge(a,b): 
    i = j = 0
    res = []
    while i < len(a) and j < len(b): 
        if a[i] > b[j]: 
            res.append(b[j])
            j += 1
        else: 
            res.append(a[i])
            i += 1

    res = res + a[i:]
    res = res + b[j:]
    return res
lst = [int(n) for n in input().split()]

print(bubble_sort(lst))
endin = len(lst) - 1
print(merge_sort(lst, 0, endin))
```

```python
#8 < notice > 
# xử lý input
s1 = input().strip()
# bỏ [[ ]] rồi tách thành các cặp
s1 = s1[2:-2]              # bỏ [[ và ]]
pairs = s1.split("], [")   # tách từng khoảng

intervals = []
for p in pairs:
    l, r = map(int, p.split(","))
    intervals.append((l, r))

# đọc dòng queries
s2 = input().strip()
s2 = s2[1:-1]              # bỏ [ ]
queries = list(map(int, s2.split(",")))
res = []
for i in queries: 
    temp = 1e10
    for j in intervals: 
       if j[1] >= i and j[0] <= i: 
            temp = min(temp, j[1] - j[0])
    if temp == 1e10: 
        res.append(-1)
    else: 
        res.append(temp)
print(res)
```

```python
# 11
def check(s1,s2, start): 
    i = start 
    j = 0
    while j < len(s2): 
        if s1[i] != s2[j]:
            return 0
        i += 1 
        j += 1
    return 1

res = 0
s = str(input())
tag = str(input())
for i in range(len(s)): 
    if s[i] == tag[0]: 
        res += check(s,tag, i)
print(res)
```
