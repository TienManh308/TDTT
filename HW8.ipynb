```python
#1
class Cylinder:
    def __init__(self, banKinh, chieuCao):
        self.banKinh = banKinh
        self.chieuCao = chieuCao
    def dienTichBeMat(self):    
        h = self.chieuCao
        r = self.banKinh
        pi = 3.14
        return round(pi*r**2*2 + 2*pi*r*h, 2)
    def theTich(self):
        h = self.chieuCao
        r = self.banKinh
        pi = 3.14

        return round(pi*r**2*h, 2)
        
r,h = map(float, input().split())
c1 = Cylinder(r,h)
print(c1.dienTichBeMat())
print(c1.theTich())
```

```python
#4
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2

    def product(self):
        return self.num1 * self.num2

    def divide(self):
        if self.num2 == 0:
            return "Khong the chia cho 0"
        return self.num1 / self.num2

    def power(self):
        return self.num1 ** self.num2

    def mod(self):
        if self.num2 == 0:
            return "Khong the chia cho 0"
        return self.num1 % self.num2

    def set_numbers(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

try:
    
    print("Nhap 2 so can tinh:")
    a = float(input("So thu nhat: "))
    b = float(input("So thu hai: "))
    
    may_tinh = Calculator(a, b)

    while True:
        print("\n--- CHON PHEP TINH ---")
        print("1. Cong (add)")
        print("2. Tru (subtract)")
        print("3. Nhan (product)")
        print("4. Chia (divide)")
        print("5. Luy thua (power)")
        print("6. Chia lay du (mod)")
        print("7. Nhap lai 2 so khac (set_numbers)")
        
        chon = int(input("Nhap lua chon (1-7): "))

        if chon == 1:
            print("Ket qua:", may_tinh.add())
        elif chon == 2:
            print("Ket qua:", may_tinh.subtract())
        elif chon == 3:
            print("Ket qua:", may_tinh.product())
        elif chon == 4:
            print("Ket qua:", may_tinh.divide())
        elif chon == 5:
            print("Ket qua:", may_tinh.power())
        elif chon == 6:
            print("Ket qua:", may_tinh.mod())
        elif chon == 7:
            a = float(input("Nhap so thu nhat moi: "))
            b = float(input("Nhap so thu hai moi: "))
            may_tinh.set_numbers(a, b)
            print("Da cap nhat so moi.")

        # Kiểm tra điều kiện thoát theo đề bài
        hoi = input("Ban co muon thoat? ")
        if hoi.lower() == 'yes':
            break

except ValueError:
    print("Loi nhap lieu")
```

```python
#6
class rectangle:
    def __init__(self, w,h):
        self.h = h
        self.w = w
    def area(self):
        return self.w*self.h
    def perimeter(self):
        return 2*(self.h+self.w)
    def scale(self, k):
        self.h *= k
        self.w *= k

height, weight, k = map(float, input().split())
rec1 = rectangle(weight, height)
rec1.scale(k)
print(rec1.area(), rec1.perimeter())
        
```

```python
#7
import math 
class fraction:
    def __init__(self,num, den):
        """"mẫu số phải dương"""
        self.num = num
        self.den = den

    def rut_gon(self):
        return self.num//math.gcd(self.num, self.den), self.den//math.gcd(self.num, self.den)
    

def tinh(a,b, op, c, d):
    match op:
        case '+':
            return (math.lcm(b,d)//b*a + math.lcm(b,d)//d*c), math.lcm(b,d)
        case '-':
            return (math.lcm(b,d)//b*a - math.lcm(b,d)//d*c), math.lcm(b,d)
        case '*':
            return a*c, b*d
        case '/':
            return a*d, b*c
        
a,b,op,c,d = input().split()
x,y = tinh(int(a), int(b), op, int(c), int(d))
f1 = fraction(x,y)
r1,r2 = f1.rut_gon()
if r1 == 0:
    print('0')
else:
    print(r1,'/',r2, sep = '')
```
