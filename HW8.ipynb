```python
#1
class Cylinder:
    def __init__(self, banKinh, chieuCao):
        self.banKinh = banKinh
        self.chieuCao = chieuCao
    def dienTichBeMat(self):    
        h = self.chieuCao
        r = self.banKinh
        pi = 3.14
        return round(pi*r**2*2 + 2*pi*r*h, 2)
    def theTich(self):
        h = self.chieuCao
        r = self.banKinh
        pi = 3.14

        return round(pi*r**2*h, 2)
        
r,h = map(float, input().split())
c1 = Cylinder(r,h)
print(c1.dienTichBeMat())
print(c1.theTich())
```

```python
#2
def nhuan(year):
    if year % 400 == 0:
        return False
    elif year % 100 or year % 4 == 0: 
        return True
class date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year  = year

    def exist_day(self):
        d = self.day
        m = self.month
        y = self.year
        if m in [1,3,5,7,8,10,12]:
            if 1 <= d <= 31:
                return True
        elif m == 2:
            if nhuan(y):
                if 1 <= d <= 29:
                    return True 
            else:
                if 1 <= d<=28:
                    return True
        elif m in [4,6,9,11]
            if 1 <= d <= 30:
                return True
        return False
    
    def tommorow(self):
        d = self.day
        m = self.month
        y = self.year
        if m in [1,3,5,7,8,10,12]:
            if m == 12 and d == 31 : return 1,1,y+1
            if d == 31:
                d = 1
                m += 1
            return d,m,y
        
        elif m == 2:
            if nhuan(y):
                if d ==  29:
                    return 1, m+1, y
            else:
                if d == 28:
                    return 1, m+1, y
        if m in [4,6,9,11]:
            if d == 30:
               return 1, m+1, y
        return d+1, m, y

day, month, year = map(int, input().split('/'))
day1 = date(day, month, year)

    
```
