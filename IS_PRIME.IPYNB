```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

a,b = map(int, input().split())
sum = 0
for num in range(a, b + 1):
    if is_prime(num):
        sum += num
print(sum)
```

```python
#26
num = int(input("Nhập số A: "))

def fibo(a):
    fi1, fi2 = 1, 1
    while fi2 <= a:
        fi1, fi2 = fi2, fi1 + fi2
    return fi1

print(f"Số Fibonacci lớn nhất không vượt quá {num} là:", fibo(num))

```
